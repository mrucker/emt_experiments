```python
import coba as cb
import matplotlib.pyplot as plt

c1 = "#E69F00"
c2 = "#0072B2"
c3 = "#009E73"
c4 = "#56B4E9"
c5 = "#D55E00"

plt.rc('font',weight='normal',size=14)
```

```python
#we remove the very first interaction from environments because it includes initializiation time
r = cb.Result.from_file("../results/capacity.log.gz").filter_fin(4000).where(index={'>':1})
```

```python
plt.figure(figsize=(6,3))

r.plot_learners(labels=['25','50','100','200','300'],err='bs',out=None,xlim=(100,None))
plt.title('',loc='left')

plt.gca().get_legend().remove()
plt.xticks([])
plt.xlabel('')
plt.yticks([.50,.55,.60],['0.500','0.550','0.600'])

plt.text(0,.615,'a)')
plt.savefig('c_acc.pdf',format='pdf', bbox_inches="tight")
```

```python
plt.figure(figsize=(6,3))

r.plot_learners(y='predict_time',labels=['25','50','100','200','300'],err='bs',out=None)

plt.title('',loc='left')
plt.ylabel("Query (ms)")

plt.gca().get_legend().remove()
plt.xticks([])
plt.xlabel('')
plt.text(-75,.07,'b)')

plt.savefig('c_query.pdf',format='pdf', bbox_inches="tight")
```

```python
plt.figure(figsize=(6,3))

r.plot_learners(y='learn_time',labels=['25','50','100','200','300'],err='bs',out=None)

plt.title('',loc='left')
plt.legend(loc='lower right')
plt.ylabel("Learn (ms)")

plt.text(-75,0.0088,'c)')

legend  = dict(columnspacing=1,handletextpad=.4,handlelength=1)
handles, labels = plt.gca().get_legend_handles_labels()    
plt.legend(reversed(handles), reversed(labels),**legend, ncols=5, bbox_to_anchor=(.945,-.35))

#plt.show()
plt.savefig('c_learn.pdf',format='pdf', bbox_inches="tight")
```
