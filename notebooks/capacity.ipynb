```python
import coba as cb
import matplotlib.pyplot as plt

c1 = "#E69F00"
c2 = "#0072B2"
c3 = "#009E73"
c4 = "#56B4E9"
c5 = "#D55E00"

plt.rc('font',weight='normal',size=14)
```

```python
#we remove the very first interaction from environments because it includes initializiation time
r = cb.Result.from_file("../results/capacity.log.gz").filter_fin(4000).where(index={'>':1})
```

```python
plt.figure(figsize=(6,3.5))

r.plot_learners(labels=['c=25','c=50','c=100','c=200','c=300'],err='bs',out=None,xlim=(100,None))
plt.title('',loc='left')
plt.legend(ncols=2,loc='lower right')

#plt.show()
plt.savefig('c_acc.pdf',format='pdf', bbox_inches="tight")
```

```python
plt.figure(figsize=(6,3))

r.plot_learners(y='predict_time',labels=['c=25','c=50','c=100','c=200','c=300'],err='bs',out=None)

plt.title('',loc='left')
plt.legend(loc='lower right')
plt.ylabel("Query Time (ms)")
plt.gca().get_legend().remove()

#plt.show()
plt.savefig('c_query.pdf',format='pdf', bbox_inches="tight")
```

```python
plt.figure(figsize=(6,3))

r.plot_learners(y='learn_time',labels=['c=25','c=50','c=100','c=200','c=300'],err='bs',out=None)

plt.title('',loc='left')
plt.legend(loc='lower right')
plt.ylabel("Learn Time (ms)")
plt.legend(bbox_to_anchor=(1,1.04))

#plt.show()
plt.savefig('c_learn.pdf',format='pdf', bbox_inches="tight")
```
