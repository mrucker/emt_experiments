```python
import coba as cb
import matplotlib.pyplot as plt

c1 = "#E69F00"
c2 = "#0072B2"
c3 = "#009E73"
c4 = "#56B4E9"
c5 = "#D55E00"

plt.rc('font',weight='normal',size=14)
```

```python
#we remove the very first interaction from because it includes initializiation time
r2 = cb.Result.from_file("../outcomes/node_capacity.log.gz").filter_fin(4000)
r2 = cb.Result(r2.environments,r2.learners,r2.interactions.where(index=1,comparison='>'))
```

```python
plt.figure(figsize=(6,3.5))

r2.plot_learners(labels=['c=10','c=25','c=50','c=100','c=150'],err='se',out=None)
plt.title('',loc='left')
plt.legend(ncols=2,loc='lower right')

#plt.show()
plt.savefig('c_acc.pdf',format='pdf', bbox_inches="tight")
```

```python
plt.figure(figsize=(6,3))

r2.plot_learners(y='predict_time',labels=['c=10','c=25','c=50','c=100','c=150'],err='se',out=None)

plt.title('',loc='left')
plt.legend(loc='lower right')
plt.ylabel("Query Time (ms)")
plt.yticks([.0005,.001,.0015,.002,.0025],[.5,1,1.5,2,2.5])
plt.gca().get_legend().remove()

#plt.show()
plt.savefig('c_query.pdf',format='pdf', bbox_inches="tight")
```

```python
plt.figure(figsize=(6,3))


r2.plot_learners(y='learn_time',labels=['c=10','c=25','c=50','c=100','c=150'],err='se',out=None)

plt.title('',loc='left')
plt.legend(loc='lower right')
plt.ylabel("Learn Time (ms)")
plt.yticks([.0002,.0004,.0006,.0008],[.2,.4,.6,.8])
plt.legend(bbox_to_anchor=(1,1.04))

#plt.show()
plt.savefig('c_learn.pdf',format='pdf', bbox_inches="tight")
```
