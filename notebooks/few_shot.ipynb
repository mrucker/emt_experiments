```python
import coba as cb
import matplotlib.pyplot as plt

c1 = "#E69F00"
c2 = "#0072B2"
c3 = "#009E73"
c4 = "#56B4E9"
c5 = "#D55E00"

plt.rc('font',weight='normal',size=14)
```

```python
r1 = cb.Result.from_file("../results/few_shot.log.gz")
```

```python
plt.figure(figsize=(6,3.3))
r1.plot_learners(err='se',labels=["OAA","LOM","RT","EMT","CMT"],colors=[c1,c2,c3,c4,c5],out=None)
plt.title('',loc='left')
plt.tight_layout()
plt.legend(bbox_to_anchor=(1,1.04))
plt.ylabel('Accuracy')

plt.xlabel('$n$-shot')
plt.xticks([0,20000,40000,60000,80000,100000],["0","20","40","60","80","100"])
plt.yticks([0,.2,.4,.6,.8],["0%","20%","40%","60%","80%"])

#plt.show()
plt.savefig('aloi_acc.pdf',format='pdf', bbox_inches="tight")
```

```python
plt.figure(figsize=(6,3.3))
r1.plot_learners(err='se',labels=["OAA","LOM","RT","EMT","CMT"],colors=[c1,c2,c3,c4,c5],y='learn_time',out=None)
plt.title('',loc='left')

plt.tight_layout()
plt.legend(bbox_to_anchor=(1,1.04))

plt.yticks([0,.0005,.001,.0015],[0,.5,1,1.5])
plt.ylabel('Learn Time (ms)')

plt.xlabel('$n$-shot')
plt.xticks([0,20000,40000,60000,80000,100000],["0","20","40","60","80","100"])

#plt.show()
plt.savefig('aloi_lrn.pdf',format='pdf', bbox_inches="tight")
```
